# 1. Installation instructions

## 1.1 Installing Anaconda

To get started, you will need to download Anaconda with the latest version of Python.

To download Anaconda, you can just follow the instructions provided in the following link: [here](https://www.anaconda.com/products/individual)

Anaconda sets up two things for you: **conda** and the **root environment**.

The process looks like this: 

  * The installer installs conda first, which is the package and
environment management tool. 

  * Then, conda creates a root environment that contains two things: a certain version of Python and some basic packages.


You can install all the libraries you need in your root environment using a package manager (like pip and conda) as it will be explained in 1.3) 

You can easily find your root environment by opening Anaconda Navigator as shown in the following figure:

````
from IPython.display import Image # for showing graphs from the lectures
Image(filename='./gdrive/My Drive/Colab Notebooks/Session_2022/Python_Session/images/env.png',  width=700) 
````

However, a better way of managing your Python projects is to create an environment for each project you are working on. We will detail how to create a new environment in the following section

## 1.2 Managing Environments

Next to the root environment, you can create as many additional environments as you want. 

And the whole point is that these additional environments can contain different versions of Pythons and other packages. 

Here is a step by step guide of the process:

1. First, open the command line (or the Terminal on Linux/Mac OS X). To create a new environment named, for instance "Imperial_Course" (you can name it whatever you
like), that includes, lets say, a Python version 3.7, run: 

  `conda create --name Imperial_Course python=3.7`

2. To activate the new environment, use the following commands depending
on the operating system you have:

  * On Linux or Mac OS X:

    `conda activate Imperial_Course`

  * On Windows:

    `activate Imperial_Course`

3. The command prompt changes upon the environments activation.

  `(Imperial_Course) ~macbook$`

4. The directories of the active environments executable files are added to the system path (this means that you can now access them more easily). You can leave an environment with this command:

  * On Linux or Mac OS X:

    `conda deactivate`

  * On Windows:

    `deactivate`


Once you activate your new environment, you can use conda to install all the needed packages as seen in the previous section.


 A quick way to check if the installation succeeded is to try to open Anaconda Navigator, switch to your new environment in the Applications on drop-down menu, and then choose to open Jupyter

## 1.3 Installing Packages

You can use conda to install a few packages that cover everything we need:

* To install jupyter notebook:

  `conda install jupyter notebook`

* To install scikit-learn:

  `conda install scikit-learn`

* To install matplotlib:

  `conda install matplotlib`

* To install tensorflow

  `conda install -c conda-forge tensorflow`

* To install hmmlearn:

  `pip install hmmlearn`

* Sometimes, in order to install a package, you should start by updating pip. For instance, for TensorFlow 2:
  * First upgrade pip:
    
    `pip install --upgrade pip`
  
  * Then install <package>:

    `pip install <package>`
